### 1) Faça um programa que desenhe uma esfera inscrita em um cubo, ambos centrados na origem de R³

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget 
import mpl_toolkits.mplot3d as Axes3D
```

```python
def cubo():   
    phi = np.arange(1,10,2)*np.pi/4 #ordem dos parametros do arange(inicio, fim, passo)
    Phi, Theta = np.meshgrid(phi, phi) #Meshgrid retorna dois vetores
#Parametrizacao da superficie cubica
    x = np.cos(Phi)*np.sin(Theta) 
    y = np.sin(Phi)*np.sin(Theta)
    z = np.cos(Theta)/np.sqrt(2)
    return x,y,z


fig = plt.figure()
ax = plt.axes(projection='3d')

a = 2
b = 2
c = 2
x,y,z = cubo()


ax.plot_surface(x*a, y*b, z*c,color=[1,0,0,0.3])


r = 1 #Esfera de raio 1 
#Parametrizando a superficie em coordenadas esfericas
u, v = np.mgrid[0:2 * np.pi:30j, 0:np.pi:20j]
x = np.cos(u) * np.sin(v)  
y = np.sin(u) * np.sin(v)
z = np.cos(v)

ax.plot_surface(x, y, z,cmap=plt.cm.YlGnBu_r, color=[0,0,1,0.6])

plt.show()
```
